### Air quality

```python
tokyo_gps = {'lat': 35.6895, 'lng': 139.6917}
key = '705e51936267439bae6d6d04e7158bb7'
```

```python
features = 'breezometer_aqi,local_aqi,health_recommendations,sources_and_effects,dominant_pollutant_concentrations,pollutants_concentrations,pollutants_aqi_information'
features = 'pollutants_concentrations'
```

```python
response = requests.get(
    "https://api.breezometer.com/air-quality/v2/current-conditions?lat={}&lon={}&key={}&features={}".format(
        tokyo_gps['lat'], tokyo_gps['lng'], key, features
    )
)
response.json()
```

https://docs.breezometer.com/api-documentation/air-quality-api/v2/#examples-3

```python
dates = [
    ['2019-01-18T0:00:00', '2019-01-24T23:00:00'],
    ['2019-01-25T0:00:00', '2019-01-31T23:00:00'],
    ['2019-02-01T0:00:00', '2019-02-07T23:00:00'],
    ['2019-02-07T0:00:00', '2019-02-13T23:00:00'],
]
```

```python
pollutant_data = []
for d in dates:
    response = requests.get(
        "https://api.breezometer.com/air-quality/v2/historical/hourly?lat={}&lon={}&key={}&start_datetime={}&end_datetime={}&features={}".format(
            tokyo_gps['lat'], tokyo_gps['lng'], key, d[0], d[1], features
        )
    )
    pollutant_data += response.json()['data']
```

```python
response.text
```

```python
pollutant_data
```

```python
pollutant_data[-1]
```

```python
pollutant_conc = [{
    'datetime': p['datetime'],
    'co': p['pollutants']['co']['concentration']['value'],
    'no2': p['pollutants']['no2']['concentration']['value'],
    'o3': p['pollutants']['o3']['concentration']['value'],
    'pm10': p['pollutants']['pm10']['concentration']['value'],
    'pm25': p['pollutants']['pm25']['concentration']['value'],
    'so2': p['pollutants']['so2']['concentration']['value'],
} for p in pollutant_data]
```

```python
p_df = pd.DataFrame(pollutant_conc)
p_df['date'] = p_df.datetime.str[:10]
p_df.head()
```

```python
for date, data in p_df.groupby('date'):
    print(date)
```
