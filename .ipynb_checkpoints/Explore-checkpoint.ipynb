```python
from scipy.stats import norm
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def path_brownian(y_init=0.0, delta=0.25, n=8*60):
    y = []
    x = y_init
    for k in range(n):
        x = x + norm.rvs(scale=delta**2)
        y.append(x)
    return y
```

```python
n = 8*60
x = list(range(n))
for i in range(5):
    y = path_brownian(y_init=20, delta=0.2, n=n)
    plt.plot(x, y)
```
